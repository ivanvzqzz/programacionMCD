```python
import pandas as pd
```

```python
import requests

url = "http://www.dgis.salud.gob.mx/descargas/datosabiertos/nacimientos/sinac_2023.zip?V=2024.05.14"
response = requests.get(url)
with open("sinac_2023.zip", "wb") as f:
    f.write(response.content)
```

```python
import zipfile

with zipfile.ZipFile("sinac_2023.zip", "r") as zip_ref:
    zip_ref.extractall("sinac_2023")
```

```python
df = pd.read_csv("sinac_2023/Nacimientos_2023.csv")
df
```

```python
print(df.dtypes)
```

Este dataframe es un ejemplo de como algunas veces los datos capturados no tienen la forma bonita que deseamos o que estamos acostumbrados, ya que casi todas las columnas son del tipo `int64`. Aún así, veamos qué significa cada una de ellas.  

```python
df.NACIOEXTRANJERO.unique()
```

Notamos primero que la columna `NACIOEXTRANJERO` toma valores 1 o 2, entonces haciendo la prueba, vemos que el 2 corresponde a nacimiento nacional y 1 corresponde a nacimiento en el extranjero.

```python
df_nacionales = df[df.NACIOEXTRANJERO == 2]
df_nacionales
```

```python
df_nacionales.ENTIDADNACIMIENTO.unique()
```

Aquí notamos un error, México solo tiene 32 entidades federativas, entonces algo raro hay aquí. Contemos las ocurrencias en cada entidad.

```python
df_nacionales.ENTIDADNACIMIENTO.value_counts()
```

Los valores de la cola contienen muy pocos datos, por lo que lo más probable es que no hayan sido correctamente capturados. Así, vamos a eliminarlos de nuestro dataframe.

```python
entidades_no_validas = [0, 88, 99]
df = df_nacionales[~df_nacionales.ENTIDADNACIMIENTO.isin(entidades_no_validas)]
df
```

```python
df = df.sort_values(by=['ENTIDADNACIMIENTO'])
df
```

Ya está tomando sentido nuestra basaecita, la información que reemplazamos tiene sentido con la realidad. Veremos qué más podemos hacer.

¿Quien fue la mamá mayor del estado de Sonora en el año 2023?

```python
# Transformamos la fecha de object a date:

df_sonora['FECHANACIMIENTOMADRE'] = pd.to_datetime(df_sonora['FECHANACIMIENTOMADRE'], format='%d/%m/%Y')
```

```python
edad_maxima = df_sonora.loc[df_sonora.FECHANACIMIENTOMADRE.idxmin()]
edad_maxima[['FECHANACIMIENTOMADRE', 'FECHANACIMIENTO', 'FECHACERTIFICADO']]
```

¿Podremos hallar a la menor?

```python
edad_minima = df_sonora.loc[df_sonora.FECHANACIMIENTOMADRE.idxmax()]
edad_minima[['FECHANACIMIENTOMADRE', 'FECHANACIMIENTO', 'FECHACERTIFICADO']]
```
